```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns 

#Read the dataset into a DataFrame 
recent_grads = pd.read_csv('recent-grads.csv')
```

```python
# group the data by major and calculate the median unemployment rate
unemployment_rate = round(recent_grads.groupby('Major')['Unemployment_rate'].median() * 100, 2)

# sort the values in descending order and display the top 5 majors with the highest median unemployment rate
highest_unemployment_rate = unemployment_rate.sort_values(ascending=False).head(5)
print("Majors with the highest median unemployment rate:")
print(highest_unemployment_rate)

# plot the median unemployment rate for majors with the highest median unemployment rate as a pie chart
plt.figure(figsize=(10,10))
plt.pie(highest_unemployment_rate.values, labels=highest_unemployment_rate.index, autopct='%1.1f%%')
plt.title('Majors with the Highest Median Unemployment Rate')
plt.show()

# sort the values in ascending order and display the top 5 majors with the lowest median unemployment rate
lowest_unemployment_rate = unemployment_rate.sort_values(ascending=True).head(5)
print("\nMajors with the lowest median unemployment rate:")
print(lowest_unemployment_rate)
```
