```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns 

#Read the dataset into a DataFrame 
recent_grads = pd.read_csv('recent-grads.csv')
```

```python
#What majors have most full-time year round jobs?

# sort the dataframe by full-time year round employment in descending order and select the top 5 rows
top_5_full_time = recent_grads.sort_values(by='Full_time_year_round', ascending=False).head(5)

# print the major and full-time year round employment for the top 5 rows
print("\nTop 5 majors with most full-time year round jobs:")
for index, row in top_5_full_time.iterrows():
    print(row['Major'])
```

```python
# create a dataframe with the top 5 majors with most full-time year round jobs
top_5_full_time = recent_grads.sort_values(by='Full_time_year_round', ascending=False).head(5)

# create a bar plot comparing the number of men and women in the top 5 majors with most full-time year round jobs
top_5_full_time[['Major', 'Men', 'Women']].set_index('Major').plot(kind='bar', stacked=True, figsize=(11,7))

# set the title and labels for the plot
plt.title('Distribution of Men and Women in Majors with Most Full-time Year Round Jobs')
plt.xlabel('Major')
plt.ylabel('Number of Graduates')

# display the plot
plt.show()
```

```python
# create a bar plot comparing the median, p25th and p75th salaries for the top 5 majors with most full-time year round jobs
top_5_full_time[['Major', 'Median', 'P25th', 'P75th']].set_index('Major').plot(kind='bar', figsize=(11,7))

# set the title and labels for the plot
plt.title('Median, P25th and P75th Salaries for Top 5 Majors with Most Full-time Year Round Jobs')
plt.xlabel('Major')
plt.ylabel('Salary')

# display the plot
plt.show()
```
